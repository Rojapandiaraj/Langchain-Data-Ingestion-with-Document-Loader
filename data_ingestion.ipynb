```python
## Web Based Loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader=WebBaseLoader(web_paths=("https://github.com/Rojapandiaraj/LLM-Gemini-Basics",),
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                         class_=("post-title","post-content","post-header")
                     ))
                     )
```

```python
loader.load()
```

```python
## Arxiv
from langchain_community.document_loaders import ArxivLoader
docs = ArxivLoader(query="1706.03762v7",load_max_docs=2).load()
len(docs)
docs
```
